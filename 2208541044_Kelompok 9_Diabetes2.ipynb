### Normalisasi Data

Normalisasi data digunakan untuk mengubah data asli menjadi bentuk yang lebih efisien dan memiliki struktur yang bagus. Dimana data berada dalam rentang yang lebih kecil, seperti -1 hingga 1 atau 0 hingga 1. Sehingga data yang sudah di normalisasi memungkinkan untuk dianalisis dengan lebih mudah menggunakan metode tertentu.

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
d = scaler.fit_transform(data)
data_normalisasi = pd.DataFrame(d, columns = data.columns)
data_normalisasi.head(10)
```

### Menentukan data predictor dan data target

Menjadikan X sebagai data predictor dan y sebagai data target.

```python
X = data_normalisasi.drop(["Diabetes_binary"], axis = 1)
y = data_normalisasi["Diabetes_binary"]
```

```python
data_normalisasi['Diabetes_binary'].hist()
```

```python
X.head(10)
```

```python
y.head(10)
```

### Split dan Test Data

Memisahkan data training dan data test dari data predictor dan data target.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
y_train.hist()
```

### Balancing Data menggunakan Random Over Sampler

Balancing data dilakukan karena adanya data imbalance atau data yang tidak seimbang. Perlu dilakukan balancing data agar tidak terjadi kekeliruan dalam melakukan klasifikasi model. Di sini akan digunakan Random Over Sampler untuk mengatasi imbalanced data, dimana data minor akan melalui resample sampai data minor menjadi seimbang dengan data mayor.

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
X_os, y_os = RandomOverSampler().fit_resample(X_train, y_train)
```

```python
y_os.hist()
```

## Predictive Modeling

Data modelling digunakan untuk mengetahui metode yang akan digunakan pada eksekusi data. Karena data yang digunakan di sini sangatlah besar, maka beberapa metode mungkin membutuhkan waktu yang lama untuk mengeksekusi. Sehingga, kami hanya akan menggunakan beberapa metode saja, yaitu:
- Logistic Regression
- K-Nearest Neighbors (KNN)

### Logistic Regression

Logistic Regression adalah metode statiska yang digunakan untuk memodelkan hubungan antara variabel independen (biasanya dalam bentuk kategori atau biner) dengan variabel dependen yang juga dalam bentuk biner.

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr = LogisticRegression()
```

```python
##Dengan Over Sampling
model_lr.fit(X_os, y_os)
```

```python
model_lr.score(X_test, y_test)
```

```python
##Tanpa Over Sampling
model_lr.fit(X_train, y_train)
```

```python
model_lr.score(X_test, y_test)
```

### KFold Logistic Regression

KFold adalah teknik yang digunakan untuk mengecek kevalidasian/keakuratan dari suatu model. Disini akan dicek keakuratan dari model berdasarkan metode Logistic Regression.

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits = 10, shuffle = True, random_state = 0)
```

```python
score_lr = cross_val_score(model_lr, X, y, cv = k_fold, n_jobs = 1, scoring = 'accuracy')
print(score_lr)
```

```python
score_lr.mean() * 100
```

##### Metode Regresi Logistik memiliki tingkat akurasi sebesar 86.33% setelah dievaluasi menggunakan KFold. Artinya model ini sangat baik dalam memodelkan data diabates2.

### K-Nearest NeighBors (KNN)

K-Nearest Neighbors (KNN) adalah sebuah metode yang digunakan untuk mengklasifikasikan suatu objek berdasarkan kemiripin antara data baru dengan sejumlah data pada lokasi terdekat yang telah tersedia.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
##Dengan Over Sampling
model_knn = KNeighborsClassifier(n_neighbors = 3)
model_knn.fit(X_os, y_os)
```

```python
model_knn.score(X_test, y_test)
```

```python
##Tanpa Over Sampling
model_knn = KNeighborsClassifier(n_neighbors = 3)
model_knn.fit(X_train, y_train)
```

```python
model_knn.score(X_test, y_test)
```

### KFold KNN

KFold adalah teknik yang digunakan untuk mengecek kevalidasian/keakuratan dari suatu model. Disini akan dicek keakuratan dari model berdasarkan metode K_Nearest Neighbors (KNN).

```python
score_knn = cross_val_score(model_knn, X, y, cv = k_fold, n_jobs = 1, scoring = 'accuracy')
print(score_knn)
```

```python
score_knn.mean() * 100
```

##### Metode KNN memiliki tingkat akurasi sebesar 82.68% setelah dievaluasi menggunakan KFold. Artinya model ini baik dalam memodelkan data diabates2.
